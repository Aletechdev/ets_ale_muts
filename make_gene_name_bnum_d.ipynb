```python
gene_df = pd.read_csv(
    "./data/regulondb10_9/gene.txt", sep="\t", comment='#', header=None)
gene_df.columns = [
    "GENE_ID",
    "GENE_NAME",
    "GENE_POSLEFT",
    "GENE_POSRIGHT",
    "GENE_STRAND",
    "GENE_SEQUENCE",
    "GC_CONTENT",
    "CRI_SCORE",
    "GENE_NOTE",
    "GENE_INTERNAL_COMMENT",
    "KEY_ID_ORG",
    "GENE_TYPE"
]
gene_df
```

```python
geno_feat_synonym_df = pd.read_csv(
    "./data/regulondb10_9/object_synonym.txt",
    sep="\t",
    comment='#',
    header=None,
    quoting=3
)
geno_feat_synonym_df.columns = ["OBJECT_ID", "OBJECT_SYNONYM_NAME", "OS_INTERNAL_COMMENT", "KEY_ID_ORG"]
geno_feat_synonym_df
```

```python
from mutil.gene import get_gene_bnum
```

```python
gene_bnum_d = dict()
for gene_names in enzyme_gene_name_d.values():
    for gn in gene_names:
        gene_bnum_d[gn] = ''
        regdb_df = gene_df[gene_df["GENE_NAME"]==gn]
        if len(regdb_df) > 0:
            regdb_id = regdb_df.iloc[0]["GENE_ID"]
            bnum = get_gene_bnum(regdb_id, geno_feat_synonym_df)
            gene_bnum_d[gn] = bnum
gene_bnum_d
```
